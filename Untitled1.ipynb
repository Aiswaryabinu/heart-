```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble  import RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=x_train.columns,
          class_names=['0', '1'])
plt.show()
```

```python
print("Train Accuracy:", clf.score(x_train, y_train))
print("Test Accuracy:", clf.score(x_test, y_test))
```

```python
#tree with controlledd depths
clf=DecisionTreeClassifier(max_depth=3)
clf.fit(x_train,y_train)
```

```python
print("Train Accuracy:", clf.score(x_train, y_train))
print("Test Accuracy:", clf.score(x_test, y_test))
```

```python
ran=RandomForestClassifier()
ran.fit(x_train,y_train)
```

```python
#for the test data
print("Random forest accuracy :",ran.score(x_test,y_test))
```

```python
#intepret feature importance
feature_importance=ran.feature_importances_
print(feature_importance)
```

```python
#Evaluate using cross validation

scores=cross_val_score(ran,x,y,cv=5)
print(scores)
```
